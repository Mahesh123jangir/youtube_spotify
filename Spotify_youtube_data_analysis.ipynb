```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Spotify_Youtube.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
url_cols = ['Unnamed: 0', 'Url_spotify', 'Uri', 'Url_youtube', 'Title', 'Description']
df.drop(url_cols, axis=1, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
album_type_count = df['Album_type'].value_counts()
print(album_type_count)
```

```python
labels = album_type_count.index.tolist()
sizes = album_type_count.values.tolist()
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

plt.title('Album Types')
plt.legend(labels, loc=0)

plt.show()
```

```python
#top 5 track by views
```

```python
top5=df.sort_values('Views',ascending=False).head(5).loc[:,['Track','Views','Artist']]
```

```python
top5.set_index('Artist')
```

```python
top5.set_index('Artist').plot.pie(y='Views',autopct='%1.0f%%').legend(bbox_to_anchor=(1.5, 1))
```

```python
top5.plot.barh(x='Track',y='Views')
plt.xlabel('Views')
plt.show()
```

```python
# Top 10 songs based on views
top10_views = df.nlargest(10, 'Views')

# Top 10 songs based on comments
top10_comments = df.nlargest(10, 'Comments')

# Top 10 songs based on likes
top10_likes = df.nlargest(10, 'Likes')
```

```python

# Top 10 songs with most views
top10_views.plot(kind='barh', x='Track', y='Views')
plt.title('Top 10 songs with most views')
plt.show()
```

```python
# Top 10 songs with most likes
top10_likes.plot(kind='barh', x='Track', y='Likes')
plt.title('Top 10 songs with most likes')
plt.show()
```

```python
# Top 10 songs with most comments
top10_comments.plot(kind='barh', x='Track', y='Comments')
plt.title('Top 10 songs with most comments')
plt.show()
```

```python
# top 10 Artist in youtube


df_views = df.groupby('Artist')['Views'].sum().sort_values(ascending=False)[:10]


df_views.plot(kind='bar')

plt.title('Top 10 Artists on YouTube')
plt.xlabel('Artist')
plt.ylabel('Views')

plt.show()
```

```python
# top 10 spotify

df_streams = df.groupby('Artist')['Stream'].sum().sort_values(ascending=False)[:10]


df_streams.plot(kind='bar')

plt.title('Top 10 Artists on Spotify')
plt.xlabel('Artist')
plt.ylabel('Streams')

plt.show()
```
